```python
import torch
from torch import nn
```

```python
x = torch.randn(100, 3)
layer = nn.Linear(3, 5)

print(layer(x).shape)
print(layer.weight)
print(layer.bias)
```

```python
x = torch.randn(2, 5)
layer = nn.ReLU()

print(x)
print(layer(x))
```

```python
layer = nn.BatchNorm1d(3)
print(layer.weight)
print(layer.bias)

x = torch.randn(5,3)
print(x)
print(layer(x))
```

```python
layer = nn.BatchNorm2d(3)
print(layer.weight)
print(layer.bias)

x = torch.randn(5,3,32,32)
print(layer(x).shape)
```

```python
x = torch.randn(3, 7)
drop = nn.Dropout(p = 0.1)
print(x)
print(drop(x) * (1 - 0.3))

drop.eval()
print(drop(x))
```

```python
class sample_model(nn.Module):
  def __init__(self):
    super().__init__()
    self.drop_layer = nn.Sequential(
        nn.Linear(5, 7),
        nn.ReLU(),
        nn.Dropout(p = 0.3)
    )

  def forward(self, x):
    x = self.drop_layer(x)
    return x
```

```python
model = sample_model()
model.train()
```

```python
x = torch.randn(3,5)
print(model(x))
```

```python
model.eval()
print(model(x))
```
